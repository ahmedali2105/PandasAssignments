```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
state2 = pd.read_csv("states2.csv")
state2
```

```python
state2 = pd.read_csv("states2.csv").drop("Unnamed: 0",axis=1)
```

```python
state2
```

```python
print(state2.head())  # Check first few rows
print(state2.columns)  # Check column names
print(state2.dtypes)  # Check data types
```

```python
state2 = state2.drop_duplicates()
```

```python
state2
```

# 1. How does the average income vary across different states?

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='State', y='Income', data=state2.sort_values('Income', ascending=False))
plt.xticks(rotation=90)
plt.title('Average Income by State')
plt.show()
```

# 2. What is the distribution of the Hispanic population percentage across states?


```python
plt.figure(figsize=(8, 6))
sns.histplot(state2['Hispanic'], bins=20, kde=True)
plt.title('Distribution of Hispanic Population Percentage')
plt.show()
```

# 3. How does the proportion of women compare to men in each state?


```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Men', y='Women', data=state2)
plt.title('Proportion of Women vs. Men by State')
plt.show()
```

# 4. Which states have the highest and lowest total populations?

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='State', y='TotalPop', data=state2.sort_values('TotalPop', ascending=False))
plt.xticks(rotation=90)
plt.title('Total Population by State')
plt.show()
```

# 5. What is the racial composition of states with the highest income?


```python
top_income_states = state2.nlargest(5, 'Income')
top_income_states.set_index('State')[['Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']].plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Racial Composition of States with Highest Income')
plt.show()
```

# 6. How does the Black population percentage correlate with average income?


```python
sns.scatterplot(x='Black', y='Income', data=state2)
plt.title('Black Population Percentage vs. Average Income')
plt.show()
```

# 7. Which states have the highest percentage of Asian population?


```python
plt.figure(figsize=(10, 6))
sns.barplot(x='State', y='Asian', data=state2.sort_values('Asian', ascending=False))
plt.xticks(rotation=90)
plt.title('Asian Population Percentage by State')
plt.show()
```

# 8. What is the relationship between the Native population percentage and total population?


```python
sns.scatterplot(x='Native', y='TotalPop', data=state2)
plt.title('Native Population Percentage vs. Total Population')
plt.show()
```

# 9. How does the Pacific population percentage vary acrmoss states?

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y='Pacific', data=state2)
plt.title('Pacific Population Percentage Across States')
plt.show()
```

# 11. What is the income distribution across states with high White population percentages?


```python
sns.scatterplot(x='White', y='Income', data=state2)
plt.title('White Population Percentage vs. Average Income')
plt.show()
```

# 12. How does the Hispanic population percentage compare to the White population percentage?


```python
sns.scatterplot(x='Hispanic', y='White', data=state2)
plt.title('Hispanic vs. White Population Percentage')
plt.show()
```

# 13. Which states have the highest and lowest gender disparities?


```python
state2['Gender_Disparity'] = abs(state2['Men'] - state2['Women'])
plt.figure(figsize=(10, 6))
sns.barplot(x='State', y='Gender_Disparity', data=state2.sort_values('Gender_Disparity', ascending=False))
plt.xticks(rotation=90)
plt.title('Gender Disparity by State')
plt.show()
```

# 14. What is the correlation between total population and average income?


```python
sns.scatterplot(x='TotalPop', y='Income', data=state2)
plt.title('Total Population vs. Average Income')
plt.show()
```
